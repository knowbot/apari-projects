**Let's define our objective functions**

```python
def objective_eggholder(x1,x2):
  ## Find the global minimum [f(512, 404.2319) = -959.6407]
  a=np.sqrt(np.fabs(x2+x1/2+47))
  b=np.sqrt(np.fabs(x1-(x2+47)))
  c=-(x2+47)*np.sin(a)-x1*np.sin(b)
  return c

def objective_levi_n13(x1, x2):
  ## Find the global minimum [f(1, 1) = 0]
  c = (np.sin(3*np.pi*x1) ** 2) + ((x1-1)**2) * (1+np.sin(3*np.pi*x2) ** 2) + \
    ((x2-1)**2) * (1+np.sin(2*np.pi*x2) ** 2)
  return c
```

Plotting the objective function

```python
def plot_objective(f, x1, x2):
    fig=plt.figure()
    ax=plt.axes(projection='3d')
    X1,X2=np.meshgrid(x1,x2)
    ax.plot_surface(X1,X2,f(X1,X2),color='red')
    ax.plot_wireframe(X1,X2,f(X1,X2),ccount=15,rcount=15,color='orange',alpha=0.8)
    ax.set_xlabel('x1')
    ax.set_ylabel('x2')
    ax.set_zlabel('f(x1,x2)')
    
def plot_solution(f, point, value, x1, x2):
    fig=plt.figure(figsize=(6, 8), dpi=100)
    ax=plt.axes(projection='3d')
    X1,X2=np.meshgrid(x1,x2)
    ax.scatter3D(point[0], point[1], value, 'blue', s=30)
#     ax.plot_surface(X1,X2,f(X1,X2),color='red',alpha=0.1)
    ax.plot_wireframe(X1,X2,f(X1,X2),ccount=15,rcount=15,color='red',alpha=0.4)
    ax.set_xlabel('x1')
    ax.set_ylabel('x2')
    ax.set_zlabel('f(x1,x2)')


plot_objective(objective_levi_n13, np.linspace(-5,5,100), np.linspace(-5,5,100))
plot_objective(objective_eggholder, np.linspace(-512,512,100), np.linspace(-512,512,100))
```

The main loop of the algorithm. Record the best solutions in each loop and print it out to see the progress

```python
# Put your code here
import collections
def get_2d_point(bounds):
    return bounds[:, 0] + np.random.rand(len(bounds)) * (bounds[:, 1] - bounds[:, 0])

def get_step(bounds, step):
    return np.random.randn(len(bounds)) * step


def check_bounds(point, bounds):
    clipped_point = [np.clip(point[i], bounds[i, 0], bounds[i, 1]) for i in range(len(bounds))]
    return clipped_point

# simulated annealing
def simulated_annealing(function, bounds, iterations, step_size, temperature):
    history = collections.deque([])
    curr_point = get_2d_point(bounds)
    curr_sol = function(curr_point[0], curr_point[1])
    best_point, best_sol = curr_point, curr_sol
    for i in range(iterations):
        new_point = check_bounds(curr_point + get_step(bounds, step_size), bounds)
        new_sol = function(new_point[0], new_point[1])
        if new_sol < best_sol:
            best_point, best_sol = new_point, new_sol
        diff = new_sol - curr_sol
        t = temperature / float(i + 1)
        mt = np.exp(-diff / t)
        if (diff < 0) or np.random.rand() < mt:
            curr_point, curr_sol = new_point, new_sol
        history.append(best_sol)
    return best_point, best_sol, history

def differential_evolution(function, bounds, iterations, population_size, m_factor, c_rate):
    # initialize random population
    history = collections.deque([])
    population = bounds[:, 0] + (np.random.rand(population_size, len(bounds)) * (bounds[:, 1] - bounds[: ,0]))
    solutions = [function(ind[0], ind[1]) for ind in population]
    best_sol = np.min(solutions)
    best_ind = population[np.argmin(solutions)]
    curr_sol = best_sol
    for i in range(iterations):
        for j in range(population_size):
            candidates = [ind for ind in range(population_size) if ind != j]
            x = population[np.random.choice(candidates, 3, replace = False)]
            mutated = check_bounds(mutation(x, m_factor), bounds)
            crossover_ind = crossover(mutated, population[j], len(bounds), c_rate)
            target_sol = function(population[j][0], population[j][1])
            crossover_sol = function(crossover_ind[0], crossover_ind[1])
            if crossover_sol < target_sol: 
                population[j] = crossover_ind
                solutions[j] = crossover_sol
        best_sol = np.min(solutions)
        if best_sol < curr_sol:
            best_ind = population[np.argmin(solutions)]
            curr_sol = best_sol
        history.append(best_sol)
    return best_ind, best_sol, history

def mutation(x, factor):
    return x[0] + factor * (x[1] + x[2])

def crossover(mutated, target, dimensions, c_rate):
    p = np.random.rand(dimensions)
    return [mutated[i] if p[i] < c_rate else target[i] for i in range(dimensions)]
```

Show the solutions plotted on the function to check if the solution makes sense

```python
bounds = np.asarray([(-5.0, 5.0), (-5.0, 5.0)])
# print((bounds[:, 0] + np.random.rand(len(bounds)) * (bounds[:, 1] - bounds[:, 0])))
print("Optimizing Levi13 with simulated annealing...")
point, value, history = simulated_annealing(objective_levi_n13, bounds, 5000, 0.5, 1)
print(f"Found global minimum of value {value} with coordinates ({point[0]}, {point[1]})")
plt.plot(history)
plt.xlabel("Iteration")
plt.ylabel("Solution")
plt.show()
plot_solution(objective_levi_n13, point, value, np.linspace(-5,5,100), np.linspace(-5,5,100))

```

```python
print("Optimizing Levi13 with differential evolution...")
point, value, history = differential_evolution(objective_levi_n13, bounds, 1000, 100, 0.6, 0.8)
print(f"Found global minimum of value {value} with coordinates ({point[0]}, {point[1]})")
plt.plot(history)
plt.xlabel("Iteration")
plt.ylabel("Solution")
plt.show()
plot_solution(objective_levi_n13, point, value, np.linspace(-5,5,100), np.linspace(-5,5,100))
```

```python
bounds = np.asarray([(-512.0, 512.0), (-512.0, 512.0)])
print("Optimizing Eggholder with simulated annealing...")
point, value, history = simulated_annealing(objective_eggholder, bounds, 200000, 1000, 1)
print(f"Found global minimum of value {value} with coordinates ({point[0]}, {point[1]})")
plt.plot(history)
plt.xlabel("Iteration")
plt.ylabel("Solution")
plt.show()
plot_solution(objective_eggholder, point, value, np.linspace(-512,512,100), np.linspace(-512,512,100))
```

```python
bounds = np.asarray([(-512.0, 512.0), (-512.0, 512.0)])
print("Optimizing Eggholder with differential evolution...")
point, value, history = differential_evolution(objective_eggholder, bounds, 1000, 100, 0.6, 0.8)
print(f"Found global minimum of value {value} with coordinates ({point[0]}, {point[1]})")
plt.plot(history)
plt.xlabel("Iteration")
plt.ylabel("Solution")
plt.show()
plot_solution(objective_eggholder, point, value, np.linspace(-512,512,100), np.linspace(-512,512,100))
```

Which algorithms is better? Write observations about algorithm performances, drawbacks, etc...

### Results
Simulated annealing seems to perform better when optimizing Levi13, with relatively "light" settings (in terms of conmputation) consistently getting a value in the ballpark of 1e-3, very close to the global minimum of 0, with equally accurate coordinates. When optimizing Eggholder, though, the algorithm sometimes gets "stuck" in local minima that are not necessarily close to the global minimum, and is fairly inconsistent overall. However when it does find the global minimum, it is very accurate in doing so. On the other hand, while differential evolution is more computationally expensive and slightly less accurate, it is much more consistent in finding a value close to the global minimum for both functions, especially for Eggholder. Therefore I would say that simulated annealing is better for optimizing Levi13, whereas differential evolution is better for Eggholder due to its consistency.
#### Performance
Simulated annealing is overall much faster than differential evolution on a per-iteration basis, also due to the fact that differential evolution is more expensive to compute since each iteration executes operations on the whole population. However as previousy said, with more irregular functions such as the Eggholder it is not as consistent.
#### Drawbacks
I suppose that the major drawback of simulated annealing is that when used on very irregular functions, the initial steps can lead the algorithm to get "stuck" in local minima that are far from the global minimum. On the other hand, differential evolution can be less accurate since it has a bigger randomizing factor with mutations and crossover, and is more expensive to compute.

Traveling Salesman Problem is the most famous NP-hard benchmark problem. You get a list of cities that you need to visit and return to the start position. The better route is the one with the shortest cost.

Implement an algorithm that will improve the route on the list of given cities.

```python
# Libs
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cm as cmx

import random, operator, time, itertools, math
import numpy as np
```

Some helper functions

```python
def distance(city1, city2): 
    # The Euclidean distance between two cities.
    # city1 and city2 are two dimensional vectors representing coordinates
    return np.linalg.norm(city1-city2)

def total_distance(city_list, tour):
    # The total distance between each pair of consecutive cities in the tour.
    # This is your objective function
    return sum(distance(city_list[tour[i]], city_list[tour[i-1]]) 
               for i in range(len(tour)))

c1 = np.array([0.0, 300.0])
c2 = np.array([400.0, 0.0])
distance(c1, c2)
```

```python
def generate_random_problem(num_cities=15):
    return [((np.random.rand(1,2) + 1) * 200) for _ in range(num_cities)]

def generate_predefined_problem():
    p = [np.array([[370, 260]]),
    np.array([[355, 378]]),
    np.array([[307, 369]]),
    np.array([[302, 399]]),
    np.array([[238, 265]]),
    np.array([[215, 333]]),
    np.array([[304, 282]]),
    np.array([[277, 207]]),
    np.array([[227, 323]]),
    np.array([[394, 314]]),
    np.array([[220, 226]]),
    np.array([[321, 333]]),
    np.array([[363, 252]]),
    np.array([[272, 257]]),
    np.array([[397, 341]])]
    return p

city_list = generate_predefined_problem()
```

```python
class AntColonyOptimizer(object):
    def __init__(self, nodes, ants, best, iters, alpha, beta, decay):
        self.nodes = nodes
        self.pheromone = np.ones((len(nodes), len(nodes)))
        self.idxs = range(len(nodes))
        self.ants = ants
        self.best = best
        self.iters = iters
        self.alpha = alpha
        self.beta = beta
        self.decay = decay
    
    def fit(self):
        min_path = None
        best_path = (None, np.inf)
        for i in range(self.iters):
            paths = self.explore_all()
            self.leave_pheromone(paths, self.best, min_path)
            min_path = min(paths, key = lambda x: x[1])
            print(min_path)
            best_path = min_path if (min_path[1] < best_path[1]) else best_path
            self.pheromone = self.pheromone * self.decay
        return best_path
    
    def explore_all(self):
        paths = []
        for i in range(self.ants):
            path = self.explore()
            paths.append((path, total_distance(self.nodes, path)))
        return paths
    
    def explore(self):
        first = 0
        path = [first]
        seen = set()
        seen.add(first)
        last = first
        for i in range(len(self.nodes) - 1):
            d = 0 if(len(path) < 2) else total_distance(self.nodes, path)
            next = self.choose(self.pheromone[last], d, seen)
            path.append(next)
            seen.add(next)
        return path
    
    def choose(self, pheromone, dist, seen):
        pheromone = np.copy(pheromone)
        pheromone[list(seen)] = 0
        p = pheromone ** self.alpha * ((1.0 / dist) ** self.beta) if dist > 0 else pheromone ** self.alpha
        n_p = p / p.sum()
        return np.random.choice(self.idxs, 1, p = n_p)[0]
    
    def leave_pheromone(self, paths, best, min_path):
        sorted_paths = sorted(paths, key=lambda x: x[1])
        for p, d in sorted_paths[:best]:
            for m in p:
                self.pheromone[m] += 1.0 / self.dist[m]
aco = AntColonyOptimizer(city_list, 10, 1, 100, 0.95, 1, 1)
p = aco.fit()
```
