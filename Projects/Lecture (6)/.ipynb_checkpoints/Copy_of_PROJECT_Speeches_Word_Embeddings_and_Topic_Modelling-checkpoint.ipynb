## Preprocessing the data

### Exercise 3

In this exercise we will write a function that cleans the column `speech` of the dataframe, and returns a corpus of type `List[List[str]]`. Include `"wa", "ha"` in your list of stopwords.

````python
def preprocess_speech(col: pd.Series, remove_stopwords: bool = True) -> List[List[str]]:

  preprocessed_docs = speeches["speech"]

  # Doing a simple preprocess using gensim
  preprocessed_docs = [
    simple_preprocess(t) for t in preprocessed_docs
  ]

  #clean new lines

  clean_new_lines = map(lambda x: x.replace("\n"," "), preprocessed_docs)

  # Lemmatizing
  lemmatizer = WordNetLemmatizer()
  preprocessed_docs = [
    [lemmatizer.lemmatize(token) for token in doc] for doc in preprocessed_docs
  ]

  # Removing stopwords
  if remove_stopwords:
    english_stopwords = set(nltk.corpus.stopwords.words("english") +["wa","ha"])
    preprocessed_docs = [
      [word for word in doc if word not in english_stopwords] for doc in preprocessed_docs
    ]

  return preprocessed_docs
````

````python
corpus = preprocess_speech(speeches)
````

Define a variable `corpus` with the result of applying this function to our `speeches` df.

## Getting topics

### Exercise 8

Define a `bows` variable with the result of applying `dct.doc2bow` to all documents in `corpus`.

````python
bows = [dct.doc2bow(doc) for doc in corpus]
````

### Exercise 9

Fit a Latent Dirichlet Allocation to our bags-of-words `bows` using `gensim`'s `LdaMulticore` for 5, 10, 15 and 20 topics. To get comparable results, set the `random_state` to my favorite number, 42.

````python
lda_model5 = LdaMulticore(
    bows,
    id2word=dct,
    num_topics=5,
    random_state=42,
    chunksize=100,
    passes=10,
    per_word_topics=True
)
lda_model10 = LdaMulticore(
    bows,
    id2word=dct,
    num_topics=10,
    random_state=42,
    chunksize=100,
    passes=10,
    per_word_topics=True
)
lda_model15 = LdaMulticore(
    bows,
    id2word=dct,
    num_topics=15,
    random_state=42,
    chunksize=100,
    passes=10,
    per_word_topics=True
)
lda_model20 = LdaMulticore(
    bows,
    id2word=dct,
    num_topics=20,
    random_state=42,
    chunksize=100,
    passes=10,
    per_word_topics=True
)
````

### Exercise 10

Plot a line with number of topics (5, 10, 15 and 20) in the x axis, and log-perplexity of the model in the y axis. Interpret it. What does the log-perplexity usually say?

````python
log_5topic = lda_model5.log_perplexity(bows)
log_10topic = lda_model10.log_perplexity(bows)
log_15topic = lda_model15.log_perplexity(bows)
log_20topic = lda_model20.log_perplexity(bows)
````

````python
plt.plot((5,10,15,20),[log_5topic,log_10topic,log_15topic,log_20topic])
````
