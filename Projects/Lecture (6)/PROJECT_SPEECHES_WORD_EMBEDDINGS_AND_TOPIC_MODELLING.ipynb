## Preprocessing the data

### Exercise 3

In this exercise we will write a function that cleans the column `speech` of the dataframe, and returns a corpus of type `List[List[str]]`. Include `"wa", "ha"` in your list of stopwords.

````python
def pp_speech(raw, rm_stopwords = True):
    pp_docs = raw.copy()
    pp_docs = [simple_preprocess(t) for t in pp_docs]
    map(lambda x: x.replace('\n',' '), pp_docs)
    wnl = WordNetLemmatizer()
    pp_docs = [[wnl.lemmatize(token) for token in doc] for doc in pp_docs]
    if (rm_stopwords):
       stopwords = set(nltk.corpus.stopwords.words("english") +['wa', 'ha']) 
    pp_docs = [[word for word in doc if word not in stopwords] for doc in pp_docs]
    return pp_docs
````

Define a variable `corpus` with the result of applying this function to our `speeches` df.

````python
corpus = pp_speech(speeches['speech'], True)
````

## Getting topics

### Exercise 8

Define a `bows` variable with the result of applying `dct.doc2bow` to all documents in `corpus`.

````python
bows = [dct.doc2bow(doc) for doc in corpus]
````

### Exercise 9

Fit a Latent Dirichlet Allocation to our bags-of-words `bows` using `gensim`'s `LdaMulticore` for 5, 10, 15 and 20 topics. To get comparable results, set the `random_state` to my favorite number, 42.

````python
lda5 = LdaMulticore(
    bows,
    id2word=dct,
    num_topics=5,
    random_state=42,
    chunksize=100,
    passes=10,
    per_word_topics=True
)
lda10 = LdaMulticore(
    bows,
    id2word=dct,
    num_topics=10,
    random_state=42,
    chunksize=100,
    passes=10,
    per_word_topics=True
)
lda15 = LdaMulticore(
    bows,
    id2word=dct,
    num_topics=15,
    random_state=42,
    chunksize=100,
    passes=10,
    per_word_topics=True
)
lda20 = LdaMulticore(
    bows,
    id2word=dct,
    num_topics=20,
    random_state=42,
    chunksize=100,
    passes=10,
    per_word_topics=True
)
print("Done!")
````

### Exercise 10

Plot a line with number of topics (5, 10, 15 and 20) in the x axis, and log-perplexity of the model in the y axis. Interpret it. What does the log-perplexity usually say?

````python
log_ps = [lda5.log_perplexity(bows), lda10.log_perplexity(bows), lda15.log_perplexity(bows), lda20.log_perplexity(bows)]
````

````python
plt.plot((5,10,15,20),log_ps)
````
